# Data Cleaning - Drop Useless Columns

```python
# Based on your results: 17 columns with 100% missing
print("ðŸ§¹ CLEANING DATA...")
print("="*80)

# Get columns with 100% missing
missing_pct = (df.isnull().sum() / len(df)) * 100
cols_to_drop = missing_pct[missing_pct == 100].index.tolist()

print(f"\nðŸ“Š Columns with 100% missing: {len(cols_to_drop)}")
print("Dropping:", cols_to_drop)

# Drop these columns
df_clean = df.drop(columns=cols_to_drop)

print(f"\nâœ… Before: {df.shape[1]} columns")
print(f"âœ… After: {df_clean.shape[1]} columns")
print(f"âœ… Dropped: {len(cols_to_drop)} useless columns")

# Update df
df = df_clean.copy()
```
